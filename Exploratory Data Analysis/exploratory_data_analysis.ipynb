```python
# convert boolean data to integers

data["protected"] = data["protected"].astype(int)
data["geo_enabled"] = data["geo_enabled"].astype(int)
data["verified"] = data["verified"].astype(int)
data["profile_use_background_image"] = data["profile_use_background_image"].astype(int)
data["default_profile"] = data["default_profile"].astype(int)
data["default_profile_image"] = data["default_profile_image"].astype(int)

for i in data.index:
  if data["url"][i] is None:
    data["url"][i] = 0
  else:
    data["url"][i] = 1


data.head()
```

```python
account_age = []
for i in data.index:
    account_age.append((today - data['created_at'][i]).days)

data['account_age'] = account_age
```

Looking at the distribution of the values for the plots above, we see that the values are skewed.
I will set up some transformation parameters to try and correct the skewed features.

```python
log_follower_growth_rate = []
for i in data.index:
    log_follower_growth_rate.append(data['followers_count'][i] / data['account_age'][i])

data['log_follower_growth_rate'] = log_follower_growth_rate
data.head()
```

```python
log_friends_growth_rate = []
for i in data.index:
    log_friends_growth_rate.append(data['friends_count'][i] / data['account_age'][i])

data['log_friends_growth_rate'] = log_friends_growth_rate
data.head()
```

```python
popularity = []
for i in data.index:
  popularity.append(data['followers_count'][i] / (data['followers_count'][i] + data['friends_count'][i]))

data['popularity'] = popularity
data.head()
```

```python
log_friends_follower_ratio = []
for i in data.index:
    log_friends_follower_ratio.append( data['friends_count'][i] / data['followers_count'][i])

data['log_friends_follower_ratio'] = log_friends_follower_ratio
data.head()
```

```python
logs_data = data

# remove some columns
logs_data.drop(['Unnamed: 0', 'id', 'screen_name', 'location', 'created_at', 'lang'], axis=1, inplace=True)
logs_data.head()
```

```python
logs_data.rename(columns = {'followers_count' : 'log_followers_count', 'friends_count' : 'log_friends_count', 'listed_count' : 'log_listed_count', 'favourites_count' : 'log_favourites_count', 'statuses_count' : 'log_statuses_count'}, inplace = True)

logs_data.head()
```

```python
# remove some other columns

logs_data.drop(['url', 'geo_enabled', 'profile_use_background_image', 'default_profile_image'], axis=1, inplace=True)

data.head()
```

```python
logs_data.dtypes
```

```python
# cast columns to appropriate datatypes

logs_data['log_followers_count'] = logs_data['log_followers_count'].astype(np.float64)
logs_data['log_friends_count'] = logs_data['log_friends_count'].astype(np.float64)
logs_data['log_listed_count'] = logs_data['log_listed_count'].astype(np.float64)
logs_data['log_favourites_count'] = logs_data['log_favourites_count'].astype(np.float64)
logs_data['log_statuses_count'] = logs_data['log_statuses_count'].astype(np.float64)
logs_data['description'] = logs_data['description'].astype(str).str.replace('/',' ')

descript = []
for i in logs_data.index:
  descript.append(logs_data['description'][i])

logs_data['descript'] = descript
del logs_data['description']
```

```python
for i in logs_data.index:
  if logs_data['log_followers_count'][i] == 0:
    logs_data['log_followers_count'][i] = -1
  else:
    logs_data['log_followers_count'][i] =  np.log(logs_data['log_followers_count'][i])

  if logs_data['log_friends_count'][i] == 0:
    logs_data['log_friends_count'][i] = -1
  else:
    logs_data['log_friends_count'][i] =  np.log(logs_data['log_friends_count'][i])

  if logs_data['log_listed_count'][i] == 0:
    logs_data['log_listed_count'][i] = -1
  else:
    logs_data['log_listed_count'][i] =  np.log(logs_data['log_listed_count'][i])

  if logs_data['log_favourites_count'][i] == 0:
    logs_data['log_favourites_count'][i] = -1
  else:
    logs_data['log_favourites_count'][i] =  np.log(logs_data['log_favourites_count'][i])

  if logs_data['log_statuses_count'][i] == 0:
    logs_data['log_statuses_count'][i] = -1
  else:
    logs_data['log_statuses_count'][i] =  np.log(logs_data['log_statuses_count'][i])

  if logs_data['log_follower_growth_rate'][i] == 0:
    logs_data['log_follower_growth_rate'][i] = -1
  else:
    logs_data['log_follower_growth_rate'][i] =  np.log(logs_data['log_follower_growth_rate'][i])

  if logs_data['log_friends_growth_rate'][i] == 0:
    logs_data['log_friends_growth_rate'][i] = -1
  else:
    logs_data['log_friends_growth_rate'][i] =  np.log(logs_data['log_friends_growth_rate'][i])

  if logs_data['log_friends_follower_ratio'][i] == 0:
    logs_data['log_friends_follower_ratio'][i] = -1
  else:
    logs_data['log_friends_follower_ratio'][i] =  np.log(logs_data['log_friends_follower_ratio'][i])

logs_data.head()
```

```python
bot_score = []
for i in logs_data.index:
  temp = logs_data['log_friends_follower_ratio'][i] + logs_data['log_friends_growth_rate'][i] - logs_data['log_follower_growth_rate'][i] - logs_data['log_listed_count'][i]
  if logs_data['default_profile'][i] == 1:
    temp += 5
  if len(logs_data['descript'][i]) == 0:
    temp += 5
  if logs_data['verified'][i] == 1:
    temp -= 10
  bot_score.append(temp)

logs_data['bot_score'] = bot_score
logs_data.head()
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

clean_dataset(logs_data)
```
